### str.contians()

```python
new_df = df.loc[df['Name'].str.contains('Mega')]
new_df.head()
```

```python
new_df = df.loc[~df['Name'].str.contains('Mega')]
new_df.head()
```

### str.contains() regex

- flag = re.I ensures it is case insensitive.
- If regex = True, first parameter in contains should be the regex expression.

```python
df.loc[df['Name'].str.contains('mega', flags=re.I, regex=True)].head()
```

```python
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head()
```

# Filter and Update 

## Conditional Statements

```python
cond_df = pd.read_csv('Data\pandasBasics.csv')
```

<b> If `Type 1` is equal to `Fire` then change `Type 1` to `Flamer` </b>

```python
cond_df.loc[cond_df['Type 1']=='Fire', 'Type 1'] = 'Flamer'

cond_df
```

<b>If `Total` is greater than 250 then update `Total` to 250 </b> 

```python
cond_df.loc[cond_df['Total']>250, 'Total'] = 250

cond_df
```

<b> Note: If `Total` is equal to 250 then update `Generation` and `Legendary` to new values </b> 

```python
cond_df.loc[cond_df['Total']==250, ['Generation', 'Legendary']] = ['Latest', 'Super Legendary']

cond_df
```

# Index 

- Filtering a df removes some rows with some indexes.
- When we save the df to a new csv file, it creates a new (oftern unwanted) col in the left which stores the previous index

```python
old_df = pd.read_csv("Data\pandasBasics.csv")
```

```python
filter_old_df = old_df[old_df['Total'] > 225]
```

```python
filter_old_df
```

- The above df shows there is an Unnamed col in the left which stores the index of previous df i.e. before filtering

### reset_index()

This method ensures the filtered index is removed before saving to csv file.

### 1. method 1: in-place store

- Reset index creates a new index from zero.
- Drop index drops the previous index from filter.
- In place ensures the changes happen in-place but takes some space.

```python
new_df1 = old_df[old_df['Total']>225]
new_df1.reset_index(drop=True, inplace = True)
new_df1.head()
```

### 2. method 2: explicit store

```python
new_df2 = old_df[old_df['Total']>225]
new_df2 = new_df2.reset_index(drop=True)
new_df2.head()
```
